```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('lr1607117Scatter.csv')
```

```python
plt.scatter(data1.x, data1.line1)
x = np.linspace(0, 100, 100)
y = 1.5*x + 32
y1 = x + 40
y2 = 1.4*x + 20
plt.plot(x, y, '-r', label='y=1.5x + 32')
plt.plot(x, y1, '-y', label='y=x + 40')
plt.plot(x, y2, '-g', label='y=1.2x + 20')
plt.legend(loc='upper left')
plt.show()
```
